```python
def generate_circle_matrix(size=64, radius=20):
    matrix = np.zeros((size, size))
    center = size // 2

    for i in range(size):
        for j in range(size):
            if np.sqrt((i - center) ** 2 + (j - center) ** 2) < radius:
                matrix[i, j] = 1
    return matrix

# Generacja i wyÅwietlenie macierzy
circle_matrix = generate_circle_matrix()
plt.imshow(circle_matrix, cmap='binary')
plt.title('64x64 Circle Matrix')
plt.show()
```

```python
from scipy.integrate import quad
import numpy as np

# Definicje funkcji
def f1_curve(x):
    return -x * np.exp(-x / 2)

def f2_curve(x):
    return np.sin(np.pi * x) + 2 * np.cos(2 * np.pi * x) + 3 * np.sin(2 * np.pi * x) * np.exp(-x / 2)

def f3_curve(x):
    return 2 * x * np.exp(-x) if x > 2 else 0

# Pochodne funkcji
def f1_prime(x):
    return np.exp(-x/2) * (-x/2 - 1)

def f2_prime(x):
    return np.pi * np.cos(np.pi * x) - 4 * np.pi * np.sin(2 * np.pi * x) + \
           3 * np.pi * np.cos(2 * np.pi * x) * np.exp(-x / 2) - 3 * np.sin(2 * np.pi * x) * np.exp(-x / 2) / 2

def f3_prime(x):
    return 2 * np.exp(-x) - 2 * x * np.exp(-x)

# Funkcja do obliczenia dÅugoÅci krzywej
def curve_length(f_prime, a, b):
    integrand = lambda x: np.sqrt(1 + f_prime(x) ** 2)
    length, _ = quad(integrand, a, b)
    return length

# Obliczenia dÅugoÅci krzywej dla kaÅ¼dej funkcji w przedziale [0, 10]
length_f1 = curve_length(f1_prime, 0, 10)
length_f2 = curve_length(f2_prime, 0, 10)
length_f3 = curve_length(f3_prime, 2, 10)  # f3 jest zdefiniowane dla x > 2

length_f1, length_f2, length_f3
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.signal import convolve

# Definicje funkcji
def f1(x):
    return np.sin(x)

def f2(x):
    return np.sin(1/x)

def f3(x):
    return np.sign(np.sin(8 * x))

# Definicje jÄder konwolucji
def h1(t):
    return np.where((t >= 0) & (t <= 1), 1, 0)

def h2(t):
    return np.where((t >= -0.5) & (t <= 0.5), 1 - np.abs(t) / 2, 0)

def h3(t):
    return np.where((t >= -1) & (t <= 1), 1 - np.abs(t), 0)

def h4(t):
    return np.sinc(t / np.pi)  # sinc(t) = sin(t) / t (znormalizowane)

# Funkcja do interpolacji
def interpolate_and_convolve(func, kernel, x_original, upsample_factor):
    # Generowanie punktÃ³w oryginalnych
    y_original = func(x_original)

    # Interpolacja (uÅ¼ywamy interp1d)
    x_interpolated = np.linspace(x_original.min(), x_original.max(), len(x_original) * upsample_factor)
    interpolator = interp1d(x_original, y_original, kind='linear')
    y_interpolated = interpolator(x_interpolated)

    # Konwolucja z jÄdrem
    kernel_size = len(x_interpolated)  # dopasowanie rozmiaru jÄdra do liczby punktÃ³w
    t = np.linspace(-1, 1, kernel_size)  # przedziaÅ -1 do 1
    kernel_values = kernel(t)
    y_convolved = convolve(y_interpolated, kernel_values, mode='same') / kernel_values.sum()

    return x_interpolated, y_interpolated, y_convolved

# Funkcja do obliczenia bÅÄdu MSE
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# GÅÃ³wna czÄÅÄ programu
# Parametry
N = 100
upsample_factor_4 = 4
upsample_factor_10 = 10
x_original = np.linspace(-np.pi, np.pi, N)

# Interpolacja i konwolucja dla kaÅ¼dej funkcji i kaÅ¼dego jÄdra
functions = [f1, f2, f3]
kernels = [h1, h2, h3, h4]
kernel_names = ['h1', 'h2', 'h3', 'h4']

for func in functions:
    plt.figure(figsize=(12, 8))
    for kernel, kernel_name in zip(kernels, kernel_names):
        # Interpolacja z 4x wiÄkszÄ liczbÄ punktÃ³w
        x_interp_4x, y_interp_4x, y_conv_4x = interpolate_and_convolve(func, kernel, x_original, upsample_factor_4)
        mse_4x = mean_squared_error(func(x_interp_4x), y_conv_4x)

        # Interpolacja z 10x wiÄkszÄ liczbÄ punktÃ³w
        x_interp_10x, y_interp_10x, y_conv_10x = interpolate_and_convolve(func, kernel, x_original, upsample_factor_10)
        mse_10x = mean_squared_error(func(x_interp_10x), y_conv_10x)

        # Wykresy wynikÃ³w
        plt.subplot(2, 2, kernels.index(kernel) + 1)
        plt.plot(x_original, func(x_original), 'bo', markersize=3, label='OryginaÅ')
        plt.plot(x_interp_4x, y_conv_4x, 'r-', label=f'Interpolacja 4x ({kernel_name}), MSE={mse_4x:.4f}')
        plt.plot(x_interp_10x, y_conv_10x, 'g-', label=f'Interpolacja 10x ({kernel_name}), MSE={mse_10x:.4f}')
        plt.legend()
        plt.title(f'Funkcja {func.__name__} z jÄdrem {kernel_name}')
        plt.xlabel('x')
        plt.ylabel('y')

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate, signal

# Define the functions for interpolation
def f1(x):
    return np.sin(x)

def f2(x):
    return np.sin(1 / x)

def f3(x):
    return np.sign(np.sin(8 * x))

# Define the convolution kernels
def h1(t):
    return np.where((t >= 0) & (t <= 1), 1, 0)

def h2(t):
    return np.where((t >= -0.5) & (t <= 0.5), 1, 0)

def h3(t):
    return np.where(np.abs(t) <= 1, 1 - np.abs(t), 0)

def h4(t, epsilon=1e-5):
    # Avoid division by zero using epsilon
    t = np.where(t == 0, epsilon, t)
    return np.sin(t) / t

# Generate a set of points and interpolate
N = 100  # Number of original points
x = np.linspace(-np.pi, np.pi, N)
functions = [f1, f2, f3]

# Prepare plots to visualize the original functions and interpolations
plt.figure(figsize=(15, 10))

# Apply convolution-based interpolation for each function
for i, func in enumerate(functions, start=1):
    y = func(x)

    # Perform convolution with kernel h1 as an example (similar steps can be done for h2, h3, h4)
    h = h1(np.linspace(-1, 1, len(x)))  # Kernel h1
    interpolated_y = signal.convolve(y, h, mode='same') / sum(h)  # Normalize convolution

    # Plot original and interpolated functions
    plt.subplot(3, 1, i)
    plt.plot(x, y, label=f"Original f{i}(x)", marker='o')
    plt.plot(x, interpolated_y, label=f"Interpolated f{i}(x) with h1", linestyle='--')
    plt.legend()
    plt.title(f"Function f{i}(x) Interpolation with h1 Kernel")
    plt.xlabel("x")
    plt.ylabel("y")

plt.tight_layout()
plt.show()
```

```python
from numpy import linspace, pi, sin, abs
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from PIL import Image
from os import remove


def aliasing_2d(n=5, jump=16, px=256, K=64):
    K = max(K, int(px / jump))
    '''
    n - number of wings
    jump - number of lines, that the sensor reads from each frame
    K - number of frames - for the gif and for generating the shutter_effect image

    K >= int(px/jump) - because for the lower value, the shutter_effect image wouldn't be complete
    '''
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(px / 96 - 0.1, px / 96 - 0.1))  # 1 inch == 96px
    line, = ax.plot([], [], lw=2)
    ax.set_axis_off()
    shutter_effect = Image.new('RGB', (px, px))

    theta = linspace(0, (2 * pi), 1000)

    def animate(m):
        if n % 2 == 1:
            r = sin(n * theta + m * pi / 10)
        else:
            r = abs(sin(int(n / 2) * theta + m * pi / 10)) ** 6
        '''
                the formula for graphing a shutter is different for even and odd n
                for even n, the first formula would actually graph twice as many wings than it should
                so the solution was to graph n/2 wings and then double it by using absolute value
                raising function to power, makes wings longer and more narrow
        '''
        line.set_data(theta, r)

        temp = m + int(K / 2)  # 0, 1, 2..., K-1
        if temp * jump < px:
            '''
            after the shutter_effect image is complete, it will stop adding lines to the image,
            that extend over the borders

            e.g. if jump=16, then the number of frames actually used to create a shutter_effect image is only 16 and if
            K=64, the code below would execute 48 extra times - without any effect (only slowing down the whole process)
            '''
            plt.savefig('t_frame.png')
            temp_img = Image.open('t_frame.png')
            temp_crop = temp_img.crop((0, temp * jump, px, temp * jump + jump))  # left, upper, right, lower
            shutter_effect.paste(temp_crop, (0, temp * jump))  # img_to_paste, (upper_left_column, upper_left_row)
            remove("t_frame.png")

        return line,

    animation = FuncAnimation(fig, animate, range(-int(K / 2), int(K / 2), 1), interval=50, blit=False)
    animation.save(f"rolling_shutter_{n}_wings_{K}_frames.gif", writer=PillowWriter(fps=28))
    shutter_effect.save(f"shutter_effect_{n}_wings_{jump}_lines.png")


if __name__ == '__main__':
    aliasing_2d()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Zakres i prÃ³bki
x = np.linspace(0, 2 * np.pi, 10)  # 10 punktÃ³w na przedziale [0, 2Ï]
y = np.sin(x)                      # WartoÅci funkcji sin(x)

# Tworzenie interpolacji metodÄ najbliÅ¼szego sÄsiada
nearest_interp = interp1d(x, y, kind='nearest')

# Nowy zakres do wyÅwietlenia interpolacji
x_new = np.linspace(0, 2 * np.pi, 100)  # WiÄcej punktÃ³w do wizualizacji
y_new = nearest_interp(x_new)          # Interpolowane wartoÅci

# Rysowanie wykresu
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', label='Punkty prÃ³bki', color='red')  # Punkty oryginalne
plt.plot(x_new, y_new, '-', label='Interpolacja najbliÅ¼szego sÄsiada', color='blue')  # Interpolacja
plt.plot(x_new, np.sin(x_new), '--', label='Funkcja sin(x)', color='green')  # Oryginalna funkcja
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolacja najbliÅ¼szego sÄsiada dla sin(x)')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Funkcje jÄdrowe
def h1(x):
    return np.where((x >= 0) & (x <= 1), 1, 0)

def h2(x):
    return np.where((x >= -0.5) & (x <= 0.5), 1, 0)

def h3(x):
    return np.where((x >= -1) & (x <= 1), 1 - np.abs(x), 0)

def h4(x):
    return np.sinc(x / np.pi)  # scipy/numpy sinc() normalizowane do sin(pi*x)/(pi*x)

# Dane do interpolacji
x_samples = np.linspace(0, 2 * np.pi, 10)  # Punkty prÃ³bki
y_samples = np.sin(x_samples)              # WartoÅci funkcji sin(x)

x_interpolated = np.linspace(0, 2 * np.pi, 10)  # GÄstsza siatka do interpolacji
y_interpolated = np.zeros_like(x_interpolated)

# Interpolacja z uÅ¼yciem funkcji jÄdrowych
kernel = h1  # WybÃ³r funkcji jÄdrowej (zmieÅ na h1, h2, h3, h4, aby eksperymentowaÄ)
for i, x in enumerate(x_interpolated):
    weights = kernel(x - x_samples)  # Oblicz wagi na podstawie funkcji jÄdrowej
    y_interpolated[i] = np.sum(weights * y_samples) / np.sum(weights)

# Wizualizacja
plt.figure(figsize=(10, 6))
plt.plot(x_samples, y_samples, 'o', label="Punkty prÃ³bki (dane wejÅciowe)")
plt.plot(x_interpolated, np.sin(x_interpolated), '--', label="Oryginalna sin(x)")
plt.plot(x_interpolated, y_interpolated, label="Interpolacja z jÄdrem")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolacja sin(x) z funkcjÄ jÄdrowÄ")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import metrics

def function_1(x):
    return np.sin(x)

def function_2(x):
    return np.sin(1 / x)

def function_3(x):
    return np.sign(np.sin(8 * x))

def kernel_1(x):
    return np.where(np.abs(x) < 0.5, 1, 0)

def kernel_2(x):
    return np.where((x >= 0) & (x <= 1), 1, 0)

def kernel_3(x):
    return np.where(np.abs(x) <= 1, 1 - np.abs(x), 0)

def conv_interpolate(x, y, x_new, kernel):

    y_new = np.zeros_like(x_new)
    kernel_width=np.median(np.diff(x))
    for i, x_i in enumerate(x_new):
        weights = kernel((x_i - x) / kernel_width)
        y_new[i] = np.sum(weights * y) / np.sum(weights)
    return y_new

n_samples = 100
n2_predictions = 200

x = np.linspace(-np.pi, np.pi, n_samples)

y_simple_sin = function_1(x)
y_inverted_sin = np.nan_to_num(function_2(x))
y_signum = function_3(x)

x_n2_interp = np.linspace(-np.pi, np.pi, n2_predictions)

y_simple_sin_interp = conv_interpolate(x, y_simple_sin, x_n2_interp, kernel_1)
y_inverted_sin_interp = conv_interpolate(x, y_inverted_sin, x_n2_interp, kernel_2)
y_signum_interp = conv_interpolate(x, y_signum, x_n2_interp, kernel_3)

y_simple_sin_true = function_1(x_n2_interp)
y_inverted_sin_true = function_2(x_n2_interp)
y_signum_true = function_3(x_n2_interp)

fig, axs = plt.subplots(3, 1, figsize=(10, 15))

fig.suptitle("Interpolation of Functions 1,2,3 using sample hold interpolation (Range: from -Ï to Ï, kernel: h1,h2,h3,  predictions=2*samples)")

axs[0].plot(x, y_simple_sin, label="Original - Sin(x)", color="red")
axs[0].plot(x_n2_interp, y_simple_sin_interp, "--", label="Interpolated kernel_1", color="blue")
axs[0].scatter(x_n2_interp[::4], y_simple_sin_interp[::4], color="blue", label="Interpolation Points", s=20)
axs[0].legend()

axs[1].plot(x, y_inverted_sin, label="Original - Sin(x^-1)", color="orange")
axs[1].plot(x_n2_interp, y_inverted_sin_interp, "--", label="Interpolated kernel_2", color="blue")
axs[1].scatter(x_n2_interp[::4], y_inverted_sin_interp[::4], color="blue", label="Interpolation Points", s=20)
axs[1].legend()

axs[2].plot(x, y_signum, label="Original - Signum(Sin(8x))", color="red")
axs[2].plot(x_n2_interp, y_signum_interp, "--", label="Interpolated kernel_3", color="blue")
axs[2].scatter(x_n2_interp[::4], y_signum_interp[::4], color="blue", label="Interpolation Points", s=20)
axs[2].legend()

print(f"MSE for Function Sin(x) 2*n predictions: {metrics.mean_squared_error(y_pred=y_simple_sin_interp, y_true=y_simple_sin_true):.1f}")
print(f"MSE for Function Sin(x^-1) 2*n predictions: {metrics.mean_squared_error(y_pred=y_inverted_sin_interp, y_true=y_inverted_sin_true):.1f}")
print(f"MSE for Function Sign(Sin(8x)) 2*n predictions: {metrics.mean_squared_error(y_pred=y_signum_interp, y_true=y_signum_true):.1f}")

plt.show()
```
